```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_mobile_network_operators'
# Add headers with a User-Agent to mimic a browser
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                  "AppleWebKit/537.36 (KHTML, like Gecko) "
                  "Chrome/120.0.0.0 Safari/537.36"
}

page = requests.get(url, headers=headers)

#page = requests.get(url)

soup = BeautifulSoup(page.text, 'html.parser')
print(soup)
```

```python
soup.find_all('table')[0]
```

```python
#<table class="wikitable sortable jquery-tablesorter" style="width:100%; font-size:95%;">
soup.find('table', class_ = 'wikitable sortable jquery-tablesorter')
table = soup.find_all('table')[0]
print(table)

```

```python
#now to remove th tags
world_titles = table.find_all('th')
world_titles
```

```python
#loop through the list to get the text only
world_titles = table.find_all('th')
world_table_titles = [title.text.strip() for title in world_titles]
print(world_table_titles)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns=world_table_titles)
df
```

```python
column_data = table.find_all('tr')
column_data
```

```python
#so far, we have looped through tr data, to find td which we have stripped to clean and store in a list
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    #print(individual_row_data)
    cleaned_row_data = [data.text.replace('\n', ' ').replace('\t', ' ').replace('\xa0', ' ').strip() for data in row_data]
    #print(cleaned)
    #Now we need to add this list to our dataframe
    length = len(df)
    df.loc[length] = cleaned_row_data
    #print(df)





```
