```python
import requests
from bs4 import BeautifulSoup

#URL = 'https://hotpoint.co.ke/'  # or a subpage where offers are listed
URL = 'https://hotpoint.co.ke/catalogue/fridges/side-by-side-fridges/'


headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36'
}

resp = requests.get(URL, headers=headers)
if resp.status_code != 200:
    print("Failed to fetch page:", resp.status_code)
    exit()

soup = BeautifulSoup(resp.content, 'html.parser')
```

```python
products = soup.find_all('div', class_='card-body product-card-body')

for prod in products:
    # Product name
    name_tag = prod.find('h5', class_='card-title product-card-name')
    name = name_tag.text.strip() if name_tag else "N/A"

    # Current price
    current_tag = prod.find('span', class_='stockrecord-price-current')
    current_price = current_tag.text.strip() if current_tag else "N/A"

    # Old price
    old_tag = prod.find('span', class_='stockrecord-price-old')
    old_price = old_tag.text.strip() if old_tag else "N/A"

    print(f"Product: {name}")
    print(f"  Current Price: {current_price}")
    print(f"  Original Price: {old_price}")
    print("-" * 40)
```
